### 1. Cálculo del Índice de Madurez (Relación Sólidos/Acidez)

En la industria de jugos o frutas, determinar si la materia prima está lista es clave.

```
# Datos de laboratorio
grados_brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice de madurez
indice = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice}")

if indice > 15:
    print("Estado: Fruta muy dulce / Madura")
else:
    print("Estado: Apta para procesamiento")
```

### 2. Control de Humedad en Secado

Un script simple para saber cuánta agua se ha perdido en un proceso de deshidratación.

```
peso_inicial = 100.0  # kg
humedad_inicial = 0.80  # 80%
humedad_final_deseada = 0.12  # 12%

# Fórmula de balance de masa para sólidos constantes
# Materia seca = peso_inicial * (1 - humedad_inicial)
materia_seca = peso_inicial * (1 - humedad_inicial)
peso_final = materia_seca / (1 - humedad_final_deseada)

print(f"Para llegar al 12% de humedad, el lote debe pesar: {peso_final:.2f} kg")
print(f"Agua a extraer: {peso_inicial - peso_final:.2f} kg")
```

### 3. Estimación de Rendimiento de Canal (Zootecnia/Cárnicos)

Ideal para plantas de beneficio animal para calcular eficiencia rápidamente.

```
def calcular_rendimiento(peso_vivo, peso_canal):
    rendimiento = (peso_canal / peso_vivo) * 100
    return rendimiento

# Ejemplo con un bovino
p_vivo = 500  # kg
p_canal = 265  # kg

resultado = calcular_rendimiento(p_vivo, p_canal)
print(f"El rendimiento de la canal es del {resultado}%")
```
